```python
def convert_examples(x):
    return x

gap = tfds.builder('gap')
gap_test = gap.as_dataset(split='test')
gap_test = gap_test.map(map_func=lambda x: convert_examples(x))
gap_test = gap_test.batch(1)
n_batches = gap.info.splits['test'].num_examples / 1
```

```python
def offset_char2token(input_str, char_offset, tokenizer_fn):
    str_part1 = input_str[:char_offset]
    
    tokens_in_part1 = tokenizer_fn(str_part1)
    
    return len(tokens_in_part1)
```
